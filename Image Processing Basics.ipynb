```python
import cv2
import matplotlib.pyplot as plt
```

```python
img =cv2.imread('../Image/lotus-flower.jpg')
```

```python
img.shape
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(hsv,cmap="hsv")
```

```python
## threshold
_,thresh=cv2.threshold(gray,90,255,cv2.THRESH_BINARY)
plt.imshow(thresh,cmap="gray")
```

```python
_,thresh=cv2.threshold(gray,90,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh,cmap="gray")
```

```python
_,thresh=cv2.threshold(gray,90,255,cv2.THRESH_TOZERO)
plt.imshow(thresh,cmap="gray")
```

```python
_,thresh=cv2.threshold(gray,90,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thresh,cmap="gray")
```

```python
_,thresh=cv2.threshold(gray,130,255,cv2.THRESH_TRUNC)
plt.imshow(thresh,cmap="gray")
```

```python
canny=cv2.Canny(img,40,180)
plt.imshow(canny,cmap="gray")
```
